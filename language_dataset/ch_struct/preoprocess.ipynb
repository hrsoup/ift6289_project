```python
import re
import json
import os

contentField = "paragraphs"

jsonPath = "./poetry"
saveFile = open(os.path.join("./", "poetry.txt"), "w", encoding='utf-8')

for file in os.listdir(jsonPath):
    if os.path.isfile(os.path.join(jsonPath,file)) and re.match('(.*)(\.)(json)', file) != None:
        print("processing file: %s" % file)
        poems = json.load(open(os.path.join(jsonPath,file), "r", encoding="utf8"))
        for i in range(len(poems)):
            singlePoem = poems[i]
            content = "".join(singlePoem[contentField])
            saveFile.write(content)
            if i != len(poems)-1:
                saveFile.write('\n')

saveFile.close()
```
